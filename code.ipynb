```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb_to_grayscale(image_path):
    """
    Converte uma imagem colorida para tons de cinza usando a biblioteca PIL.

    Args:
        image_path (str): Caminho para a imagem colorida.

    Returns:
        np.array: Imagem em tons de cinza como array NumPy.
    """
    image = Image.open(image_path).convert("RGB")
    image_array = np.array(image)

    # Fórmula para conversão em tons de cinza: luminância perceptual
    grayscale_array = np.dot(image_array[..., :3], [0.299, 0.587, 0.114]).astype(np.uint8)

    return grayscale_array

def grayscale_to_binary(grayscale_array, threshold=128):
    """
    Converte uma imagem em tons de cinza para binária (preto e branco) usando NumPy.

    Args:
        grayscale_array (np.array): Array NumPy de tons de cinza (0-255).
        threshold (int): Valor de limiar para binarização (padrão: 128).

    Returns:
        np.array: Imagem binária como array NumPy com valores 0 ou 255.
    """
    binary_array = np.where(grayscale_array >= threshold, 255, 0).astype(np.uint8)
    return binary_array
```

```python
# Caminho para a imagem de exemplo
image_path = "/content/lena.jpg"

# Converter para tons de cinza
grayscale_image = rgb_to_grayscale(image_path)

plt.imshow(grayscale_image, cmap="gray")
plt.title("Imagem em Tons de Cinza")
plt.axis("off")
plt.show()


# Converter para binária (preto e branco)
binary_image = grayscale_to_binary(grayscale_image)
Image.fromarray(binary_image).show()

plt.imshow(binary_image, cmap="gray")
plt.title("Imagem binarizada (preto e branco)")
plt.axis("off")
plt.show()
```
